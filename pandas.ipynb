# 3.opening a csv file from an url

```python
import requests
from io import StringIO
url="httpS://localhost:8888/edit/amy.csv"
headers={"user-agent":"Mozilla/5.0(Macintosh;Intel Mac OS X10.14;rv:66.0) Gecko/20100101 Firefox/66.0"}
req= requests.get(url,headers=headers)
data=StringIO(req.text)

pd.read_csv(data)
```
